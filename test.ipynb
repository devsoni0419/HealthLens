```python
from openai import OpenAI
import os
import cv2
import numpy as np

# ------------------------
# STEP 1: Gemini Gloss Generator
# ------------------------

client = OpenAI(
    api_key=os.getenv("GEMINI_API_KEY"),
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)

response = client.chat.completions.create(
    model="gemini-3-flash-preview",
    messages=[
        {"role": "system", "content": "Convert English to sign language gloss in uppercase without punctuation."},
        {"role": "user", "content": "I am going to school tomorrow."}
    ]
)

gloss = response.choices[0].message.content.strip()
print("Generated Gloss:", gloss)


# ------------------------
# STEP 2: Simple Gloss → Pose Frames
# ------------------------

words = gloss.split()

frame_width = 512
frame_height = 512
fps = 2
video_path = "sign_output.mp4"

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
video = cv2.VideoWriter(video_path, fourcc, fps, (frame_width, frame_height))

for word in words:
    frame = np.zeros((frame_height, frame_width, 3), dtype=np.uint8)

    # Draw stick figure body
    cv2.circle(frame, (256, 120), 40, (255, 255, 255), 2)  # head
    cv2.line(frame, (256, 160), (256, 300), (255, 255, 255), 2)  # body
    cv2.line(frame, (256, 200), (180, 250), (255, 255, 255), 2)  # left arm
    cv2.line(frame, (256, 200), (330, 250), (255, 255, 255), 2)  # right arm
    cv2.line(frame, (256, 300), (200, 400), (255, 255, 255), 2)  # left leg
    cv2.line(frame, (256, 300), (310, 400), (255, 255, 255), 2)  # right leg

    # Display gloss word
    cv2.putText(frame, word, (150, 480),
                cv2.FONT_HERSHEY_SIMPLEX, 1.5,
                (0, 255, 0), 3, cv2.LINE_AA)

    video.write(frame)

video.release()

print("Video saved as:", video_path)
```
